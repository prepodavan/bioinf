#### ответ

сделаем выравнивание между GNPAT_HUMAN и BLT3B_HUMAN

```python
%%bash
curl -s 'https://rest.uniprot.org/uniprotkb/O15228.fasta' -o 'O15228.fasta'
curl -s 'https://rest.uniprot.org/uniprotkb/A0JNW5.fasta' -o 'A0JNW5.fasta'
rm -f O15228.A0JNW5.aln.needle
needle -asequence O15228.fasta -bsequence A0JNW5.fasta -auto -outfile O15228.A0JNW5.aln.needle
cat O15228.A0JNW5.aln.needle | head -12
```

```python
import re
from IPython import display


def get_percent(data):
    return data[data.find('(')+1:data.rfind('%)')]

def parse_srspair(input_file):
    alignments = []
    long_gap_count = 0
    seq1_name, seq2_name = None, None
    alignment_data = {}
    lens = dict()
    for line in open(input_file, 'r', encoding='utf-8').readlines():
        line = line.strip()
        if line.startswith("# 1:"):
            seq1_name = line.split(":", 1)[1].strip()
        elif line.startswith("# 2:"):
            seq2_name = line.split(":", 1)[1].strip()
        elif line.startswith("# Identity:"):
            alignment_data["Identity"] = get_percent(line.split(":")[1].strip())
        elif line.startswith("# Similarity:"):
            alignment_data["Similarity"] = get_percent(line.split(":")[1].strip())
        elif line.startswith("# Gaps:"):
            alignment_data["Gaps"] = get_percent(line.split(":")[1].strip())
        elif line.startswith("# Score:"):
            alignment_data["Score"] = line.split(":")[1].strip()
        elif line.startswith("# Length:"):
            alignment_data["Length"] = line.split(":")[1].strip()
        elif line.startswith("#---"):
            alignment_data["UniProt ID 1"] = seq1_name
            alignment_data["UniProt ID 2"] = seq2_name
            alignment_data["Gaps>2"] = long_gap_count
            alignments.append(alignment_data)
            break
                
        elif (seq1_name and line.startswith(seq1_name)) or (seq2_name and line.startswith(seq2_name)):
            gaps = re.findall(r"-{3,}", line)
            if gaps:
                long_gap_count += len(list(gaps))
        
    return alignments

def alignments_to_md(alignments):
    out = f'|UniProt ID 1|UniProt ID 2|Score|% Identity|% Similarity|Gaps|Gaps>2|\n'
    out += '|---|---|---|---|---|---|---|\n'
    for alignment in alignments:
        out += f'|{alignment.get("UniProt ID 1", "")}'
        out += f'|{alignment.get("UniProt ID 2", "")}'
        out += f'|{alignment.get("Score", "")}'
        out += f'|{alignment.get("Identity", "")}'
        out += f'|{alignment.get("Similarity", "")}'
        out += f'|{alignment.get("Gaps", "")}'
        out += f'|{alignment.get("Gaps>2", "")}|\n'
    return out


display.display_markdown(alignments_to_md(parse_srspair('O15228.A0JNW5.aln.needle')), raw=True)
```

видно, что белки похожи мало друг на друга. score и similarity маленькие

## Задание 5

#### задача
Сделайте два локальных выравнивания – 

Вашего человеческого белка и ортологом и белком, выбранным в задании 4,

– с помощью программы water с параметрами по умолчанию. 

Заполните таблицу. Опишите свои наблюдения. 

По результатам заданий 4 и 5 сделайте вывод.

#### ответ



```python
%%bash
curl -s 'https://rest.uniprot.org/uniprotkb/O15228.fasta' -o 'O15228.fasta'
curl -s 'https://rest.uniprot.org/uniprotkb/Q9ES71.fasta' -o 'Q9ES71.fasta'
curl -s 'https://rest.uniprot.org/uniprotkb/A0JNW5.fasta' -o 'A0JNW5.fasta'
rm -f O15228.A0JNW5.aln.water
water -asequence O15228.fasta -bsequence A0JNW5.fasta -auto -outfile O15228.A0JNW5.aln.water
cat O15228.A0JNW5.aln.water | head -12
water -asequence Q9ES71.fasta -bsequence A0JNW5.fasta -auto -outfile Q9ES71.A0JNW5.aln.water
cat Q9ES71.A0JNW5.aln.water | head -12
water -asequence O15228.fasta -bsequence Q9ES71.fasta -auto -outfile O15228.Q9ES71.aln.water
cat O15228.Q9ES71.aln.water | head -12
```

```python
ids = ['O15228', 'Q9ES71', 'A0JNW5']


def alignments_to_md_with_coverage(alignments, seq_lens):
    lines = [
        f'|UniProt ID 1|UniProt ID 2|Score|% Identity|% Similarity|Gaps|Gaps>2|Coverage 1|Coverage 2|',
        '|---|---|---|---|---|---|---|---|---|',
    ]
    for alignment in alignments:
        seq1 = alignment.get("UniProt ID 1", "")
        seq2 = alignment.get("UniProt ID 2", "")
        out = ''
        out += f'|{seq1}'
        out += f'|{seq2}'
        out += f'|{alignment.get("Score", "")}'
        out += f'|{alignment.get("Identity", "")}'
        out += f'|{alignment.get("Similarity", "")}'
        out += f'|{alignment.get("Gaps", "")}'
        out += f'|{alignment.get("Gaps>2", "")}'
        out += f'|{int(alignment["Length"]) * 100 / seq_lens[seq1]:.2f}'
        out += f'|{int(alignment["Length"]) * 100 / seq_lens[seq2]:.2f}|'
        lines.append(out)
    return lines


fst = parse_srspair(f'{ids[0]}.{ids[2]}.aln.water')
snd = parse_srspair(f'{ids[1]}.{ids[2]}.aln.water')
thd = parse_srspair(f'{ids[0]}.{ids[1]}.aln.water')

def get_full_len(fn):
    return len(''.join([line.strip() for line in open(fn).readlines()[1:]]))

lens = {
    'GNPAT_HUMAN': get_full_len(f'{ids[0]}.fasta'),
    'BLT3B_HUMAN': get_full_len(f'{ids[2]}.fasta'),
    'GNPAT_RAT': get_full_len(f'{ids[1]}.fasta'),
}

table = alignments_to_md_with_coverage(fst, lens)
table.append(alignments_to_md_with_coverage(snd, lens)[-1])
table.append(alignments_to_md_with_coverage(thd, lens)[-1])
display.display_markdown('\n'.join(table) + '\n', raw=True)
```
